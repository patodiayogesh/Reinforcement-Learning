# Q-Learning
* In this part you will implement Q-Learning with epsilon-greedy exploration to find the optimal policy.  
  * You can experiment with fixed and adaptive learning rate and epsilon
  * Report your best result and mention the differences you observed in answer to final questions


```python
def action_policy(Q, state, epsilon = 0.2):
    
    prob = np.random.random()
    
    # Greedy Action
    greedy_action = np.argmax(Q[state,:])
    
    # Epsilon Greedy Policy
    if prob > epsilon:
        action = greedy_action
    else:
        action = np.random.choice([a for a in range(env.nA) if a != greedy_action])
    
    return action


def update_policy(policy, Q, state, epsilon = 0.2):
    
    prob = np.random.random()
    
    # Greedy Action
    greedy_action = np.argmax(Q[state,:])
    
    # Epsilon Greedy Policy
    if prob > epsilon:
        action = greedy_action
    else:
        action = np.random.choice([a for a in range(env.nA) if a != greedy_action])
    
    policy[state] = action

def create_policy(nS, nA):
    
    policy = np.zeros((nS,nA))
    for s in range(nS):
        policy[s] = np.ones(nA, dtype = float)/nA
        
    return policy
```

```python
# Initialize Running Parameters. 
# You should try to maintain the *number of episodes* and *max length of episode* at these numbers for your final results
# However feel free to modify them for debugging purposes

def q_learning(epsilon_decay = False, learn_rate_decay = False, eps = 0.2, learn_rate = 0.8):
    # Q-Learning
    
    num_episodes       = 5000
    episode_max_length = 100

    # initialize discount factor and learning rate
    gamma     = 0.95
    learnRate = learn_rate

    #create Q table matrix Q[s,a]
    Q=np.zeros([env.nS,env.nA])

    # list containing total reward of each episode (these rewards don't need to be discounted)
    reward_of_episodes = []

    # Greedy Policy Epsilon Value
    if epsilon_decay:
        epsilon = 1.0
    else:
        epsilon = eps

    #execute in episodes
    for episode in range(num_episodes):

        # always start with state 0 i.e Start
        state = env.reset()
        total_reward = 0

        for t in range(episode_max_length):

            # Decay Learn Rate
            if learn_rate_decay:
                learnRate = 1/(t+1)

            # Choose Action to be taken
            action = action_policy(Q,state,epsilon)
            newstate, reward, finished, _ = env.step(action)        

            # Update Q Values
            target = reward + gamma*np.max(Q[newstate,:])
            delta = target - Q[state][action]
            Q[state][action] = Q[state][action] + learnRate*delta
            total_reward += reward

            # Terminitate if Terminal State Reached
            if finished:
                break

            # Set Next State
            state = newstate

        reward_of_episodes.append(total_reward)

        # Decay Epsilon
        if epsilon_decay:
            epsilon = 0.01 + (1 - 0.01)*np.exp(-0.005*episode)
    
    return reward_of_episodes
```

- Construct `reward_of_episodes`, a list with the total rewards of each episode (these rewards do not need to be discounted)
- Plot the moving average of the list
- Ideally you want to achieve a reward greater than 0.7 (the problem is technically considered solved for 0.78)

```python
# You don't need to change the code here.
# you can use this function to plot the moving average of a list
from numpy import convolve, ones
def movingaverage(interval, window_size):
    window= np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'valid')
from pylab import plot
```

```python
# Plot the moving average of total undiscounted reward across episodes here

# Q Learning with Fixed Epsilon and Fixed Learn Rate
reward_of_episodes = q_learning(eps = 0.2, learn_rate = 0.8)
rm=movingaverage(reward_of_episodes, 100)
plot(rm)
```

```python
# Q Learning with Adaptive Epsilon and Fixed Learn Rate
reward_of_episodes = q_learning(epsilon_decay = True)
rm=movingaverage(reward_of_episodes, 100)
plot(rm)
```

```python
# Q Learning with Fixed Epsilon and Adaptive Learn Rate
reward_of_episodes = q_learning(learn_rate_decay = True)
rm=movingaverage(reward_of_episodes, 100)
plot(rm)
```

```python
# Q Learning with Adaptive Epsilon and Adaptive Learn Rate
reward_of_episodes = q_learning(epsilon_decay = True, learn_rate_decay = True)
rm=movingaverage(reward_of_episodes, 100)
plot(rm)
```

#  SARSA

- In the following you will implement SARSA with epsilon-greedy exploration.  
    - You can experiment with fixed and adaptive learning rate and epsilon
    - Report your best result and mention the differences you observed in answer to final questions

```python
# initialize episodic structure - again try to maintain these two parameters as they are for your final submission

def sarsa(epsilon_decay = False, learn_rate_decay = False, eps = 0.2, learn_rate = 0.8):
    # SARSA Leearning Algorithm
    
    num_episodes=5000;
    episode_max_length=100;

    # initialize discount factor, learning rate
    gamma=0.95
    learnRate=learn_rate

    # create Q table matrix Q[s,a]
    Q=np.zeros([env.nS,env.nA])

    # Create Online Policy to be Updated
    policy = np.zeros(env.nS)

    # list containing total reward of each episode
    reward_of_episodes = []

    # Greedy Policy Epsilon Value
    if epsilon_decay:
        epsilon = 1.0
    else:
        epsilon = eps

    # execute in episodes
    for episode in range(num_episodes):

        state = env.reset()
        action = action_policy(Q, state)
        total_reward = 0

        for t in range(episode_max_length):

            # Decay Learn Rate
            if learn_rate_decay:
                learnRate = 1/(t+1)

            # Observe Next State and Action
            next_state, reward, done, _ = env.step(action)
            next_action = action_policy(Q,next_state,epsilon)

            # Update Q Values
            target = reward + gamma*Q[next_state][next_action]
            delta = target - Q[state][action]
            Q[state][action] += learnRate*delta

            total_reward += reward

            # Update Policy
            update_policy(policy,Q,state,epsilon)

            if done:
                break

            # Set Next State And Action
            state, action = next_state, next_action

        reward_of_episodes.append(total_reward)

        # Decay Epsilon
        if epsilon_decay:
            epsilon = 0.01 + (1 - 0.01)*np.exp(-0.005*episode)
        
    return reward_of_episodes
```

```python
# Sarsa with Fixed Epsilon and Fixed Learn Rate
reward_of_episodes = sarsa(eps = 0.2, learn_rate = 0.8)
rm=movingaverage(reward_of_episodes, 100)
plot(rm)
```

```python
# Sarsa with Adaptive Epsilon and Fixed Learn Rate
reward_of_episodes = sarsa(epsilon_decay = True)
rm=movingaverage(reward_of_episodes, 100)
plot(rm)
```

```python
# Sarsa with Fixed Epsilon and Adaptive Learn Rate
reward_of_episodes = sarsa(learn_rate_decay = True)
rm=movingaverage(reward_of_episodes, 100)
plot(rm)
```

```python
# Sarsa with Adaptive Epsilon and Adaptive Learn Rate
reward_of_episodes = sarsa(epsilon_decay = True, learn_rate_decay = True)
rm=movingaverage(reward_of_episodes, 100)
plot(rm)
```
